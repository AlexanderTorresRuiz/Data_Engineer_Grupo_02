```python
import os
import pandas as pd
import csv

# Inicializar diccionario para almacenar DataFrames
dataframes = {}

for archivo in archivos_dict["csv"]:
    ruta_archivo = os.path.join(carpeta_entrada, archivo)  # Ruta completa

    if os.path.exists(ruta_archivo):  # Verificar si el archivo existe
        try:
            # Si el archivo contiene "google" en el nombre, omitir las dos primeras filas
            skip_rows = 2 if "google" in archivo.lower() else 0

            # Detectar delimitador autom√°ticamente
            with open(ruta_archivo, "r", encoding="utf-8") as f:
                sample = f.read(1024)
                f.seek(0)  # Volver al inicio del archivo
                dialect = csv.Sniffer().sniff(sample)

            # Leer el archivo con el delimitador detectado y saltando filas si es necesario
            df = pd.read_csv(
                ruta_archivo, 
                encoding="utf-8", 
                sep=dialect.delimiter, 
                skiprows=skip_rows,  # Aplicar el ajuste solo a archivos con "google"
                on_bad_lines="skip"
            )

            dataframes[archivo] = df  # Guardar el DataFrame
        
        except (UnicodeDecodeError, csv.Error):
            try:
                # Si falla, intentar con latin-1
                df = pd.read_csv(
                    ruta_archivo, 
                    encoding="latin-1", 
                    skiprows=skip_rows,  # Aplicar el ajuste solo a archivos con "google"
                    on_bad_lines="skip"
                )
                dataframes[archivo] = df
            except Exception as e:
                print(f"Error al leer {archivo}: {e}")
    
    else:
        print(f"Archivo no encontrado: {ruta_archivo}")
```

```python
# Leer los archivos CSV en pandas y asignarlos a variables espec√≠ficas
for archivo, df in dataframes.items():
    if "meta" in archivo.lower():
        globals()['df_meta'] = df  # Asigna a df_meta
        print(f"‚úÖ Asignado: {archivo} a df_meta")
    elif "tiktok" in archivo.lower():
        try:
            # Forzar relectura con delimitador coma y comillas
            ruta_tiktok = os.path.join(carpeta_entrada, archivo)
            df_tiktok = pd.read_csv(
                ruta_tiktok,
                encoding="utf-8",
                sep=",",             # Forzamos coma como delimitador
                quotechar='"',       # En caso tenga campos con comas dentro
                on_bad_lines="skip"
            )
            globals()['df_tiktok'] = df_tiktok
            print(f"‚úÖ Asignado: {archivo} a df_tiktok (forzado)")
        except Exception as e:
            print(f"‚ùå Error al forzar lectura de TikTok: {e}")
    elif "google" in archivo.lower():
        globals()['df_google'] = df  # Asigna a df_google
        print(f"‚úÖ Asignado: {archivo} a df_google")
    else:
        print(f"‚ö†Ô∏è No se pudo asignar el archivo: {archivo} (no coincide con 'meta', 'tiktok', o 'google')")
```
